## Importing the data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

## Displaying data

```python
df = pd.read_csv('aerofit.csv')
df
```

## *Univariate Analysis*

### Analysing the sale of the products based on category 

```python
sales_plot = sns.countplot(x=df["Product"], color='green')

sales_plot.set_title('Sales by Product')

sales_plot.xaxis.set_label_text('Product')

sales_plot.yaxis.set_label_text('Sales')
```

- From the above bar plot, it can be inferred that the base model (KP281) is the most sold product.

- Also, it  can be seen that there is a linear drop in the sales as we advance to the higher model,

### Analysis based on age of customers

```python
sns.histplot(df['Age'], kde=False, color='green')
```

```python
sns.boxplot(x=df["Age"], showmeans=True, color='yellow')
```

```python
sns.boxplot(x=df["Age"], y=df["Product"],showmeans=True, hue=df["Product"], palette=['yellow', 'red', 'lightgreen'])
```

```python
df["Age"].describe()
```

#### Inference :

- By analyzing these plots, it can be understood that, for all the three products, the customers are mostly in the age group `18-30`.

- Also, the mean and median of all the ages for all the products comes out to be almost same (around 28-29).

- There are a few outliers in the plots, suggesting that a few customers also lie in the age group `40-50`.

### Analysis based on the Gender of customers

```python
sns.countplot(x=df["Gender"], hue=df["Product"], palette=['yellow', 'red', 'lightgreen'])
```

#### Inference :

- By this graph, we can understand that the number of male and female customers are almost similar, but it can be seen that only very few units of the highest model (`KP781`) were bought by female customers.

- While, it is seen that there is no such pattern noticed in case of male customers. Instead, the highest model of the prodcut is sold more than the mid-level model. 

- One thing that is common in both the genders is that the sale of the model `KP281` is the highest and the number of units sold is also almost the same.
