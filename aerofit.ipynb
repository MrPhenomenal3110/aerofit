### Adding the required custom columns

```python
def age_group(age):
    if age <= 30:
        return '18-30'
    elif age <= 40:
        return '31-40'
    else:
        return 'greater than 40'

def education_group(education):
    if education <= 14:
        return '12-14'
    elif education <= 16:
        return '14-16'
    else:
        return 'greater than 16'

def usage_group(usage):
    if usage <= 3:
        return 'Less'
    elif usage <= 5:
        return 'Average'
    else:
        return 'Good'

def fitness_group(fitness):
    if fitness < 3:
        return 'Low'
    elif fitness == 3:
        return 'Average'
    else:
        return 'High'

def income_group(income):
    if income <= 45000:
        return 'Low'
    elif income <= 60000:
        return 'Average'
    else:
        return 'High'

def miles_group(miles):
    if miles <= 75:
        return '0-75'
    elif miles <= 150:
        return '76-150'
    else:
        return 'greater than 150'

df['Age Group'] = df['Age'].apply(age_group)
df['Education Group'] = df['Education'].apply(education_group)
df['Usage Group'] = df['Usage'].apply(usage_group)
df['Fitness Group'] = df['Fitness'].apply(fitness_group)
df['Income Group'] = df['Income'].apply(income_group)
df['Miles Group'] = df['Miles'].apply(miles_group)

df
```

### Analysing the sale of the products based on category 

```python
sales_plot = sns.countplot(x=df["Product"], color='green')

sales_plot.set_title('Sales by Product')

sales_plot.xaxis.set_label_text('Product')

sales_plot.yaxis.set_label_text('Sales')
```

```python
crosstab = pd.crosstab(index=df['Product'], columns='count')
marginal_prob_sales = crosstab / crosstab.sum()
marginal_prob_sales.columns = ['Probability']

marginal_prob_sales
```

#### Inference :

- From the bar plot, it can be inferred that the base model (KP281) is the most sold product, followed by the mid-level model (KP481) and the least sold product is the highest model (KP781).

- There is a linear drop in the sales as we advance to the higher model.
